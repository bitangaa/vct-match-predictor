Predicting an individual player's stats against a given opponent, giving over/unders for theoretical gambling game

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
```

```python
df = pd.read_csv('../Data/player_stats3.csv')
```

```python
df.head()
```

```python
# Load valid teams and filter dataframe
with open('../Data/Dictionaries/team_names.txt', encoding='utf-8') as f:
    valid_teams = set(line.strip() for line in f if line.strip())

df = df[df['team'].isin(valid_teams) & df['opponent'].isin(valid_teams)]
```

```python
df = df[(df['kills'] != 0) | (df['deaths'] != 0) | (df['assists'] != 0)]
```

```python
df.head()
```

```python
df['kda'] = (df['kills'] + df['assists']) / df['deaths'].replace(0, 1)
df['kd'] = df['kills'] / df['deaths'].replace(0, 1)
```
